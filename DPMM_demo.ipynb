```python
import numpy as np
from mixture_models import DirichletProcessMixtureModel as DPMM
import matplotlib.pyplot as plt
palettes = [
    '#D870AD', '#B377D9', '#7277D5', 
    '#4B8CDC', '#3BB1D9', '#3BBEB0', 
    '#3BB85D', '#82C250', '#B0C151', 
    '#F5BA42', '#F59B43', '#E7663F', 
    '#D94C42']
cc_ = np.array(palettes)
```

# Dirichlet Process Mixture Model

## $\alpha = 0.5$

```python
alpha_real = 0.5
dp = DPMM(alpha_real, n_dim=2, sig2=0.25)

seed = 41918
np.random.seed(seed)
data_ = dp.generate_simulated_data(100)
```

### Initialization: all points in one cluster

```python
alpha_fit = 0
# dp.max_itr = 100
dp.fit(alpha=alpha_fit, init_method=0)

plt.figure(figsize=(12,6))
ax = plt.subplot(121, aspect='equal')
ax.scatter(data_[:,0], data_[:,1], s=20, alpha=0.8)
ax.scatter(dp.cluster_parameters_[:,0], dp.cluster_parameters_[:,1], s=30, marker='s', alpha=0.8)

plt.title(r"$\alpha = {}$".format(alpha_real))
ax = plt.subplot(122, aspect='equal')
ax.scatter(data_[:,0], data_[:,1], s=20, alpha=0.8, c=cc_[dp.labels_*2])
plt.title(r"$\alpha={}$".format(alpha_fit))
plt.show()
```

```python
alpha_fit = 0.5
# dp.max_itr = 100
dp.fit(alpha=alpha_fit, init_method=0)

plt.figure(figsize=(12,6))
ax = plt.subplot(121, aspect='equal')
ax.scatter(data_[:,0], data_[:,1], s=20, alpha=0.8)
ax.scatter(dp.cluster_parameters_[:,0], dp.cluster_parameters_[:,1], s=30, marker='s', alpha=0.8)

plt.title(r"$\alpha = {}$".format(alpha_real))
ax = plt.subplot(122, aspect='equal')
ax.scatter(data_[:,0], data_[:,1], s=20, alpha=0.8, c=cc_[dp.labels_*2])

plt.title(r"$\alpha={}$".format(alpha_fit))
plt.show()
```
